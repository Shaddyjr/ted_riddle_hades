```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# number of people in line
n = 100
dead  = [0] * n
alive = [0] * n
```

```python
n_simulations = 10_000

for _ in range(n_simulations):
    line_of_people = list(range(n))
    
    # cerberus eating the line
    while line_of_people:
        bite_pos = np.random.randint(3)
        try:
            if bite_pos == 0:
                first_person_i = line_of_people.pop(0)
                alive[first_person_i] += 1
            elif bite_pos == 1:
                first_person_i = line_of_people.pop(0)
                dead[first_person_i] += 1

                second_person_i = line_of_people.pop(0)
                alive[second_person_i] += 1
            else:
                first_person_i = line_of_people.pop(0)
                dead[first_person_i] += 1

                second_person_i = line_of_people.pop(0)
                dead[second_person_i] += 1

                third_person_i = line_of_people.pop(0)
                alive[third_person_i] += 1
        except:
            pass
```

```python
sns.lineplot(range(n), np.array(alive) / n_simulations)
plt.title("Alive percentage (%) by index position")
plt.xlabel("index position")
plt.ylabel("Percentage alive");
```
